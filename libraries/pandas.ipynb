```python
#operations in pandas
#1 loading data
df=pd.read_csv("data.csv")
print(df.head())
print()
#2 veiwing and exploring data
print(df.info())
print()
#3 handling missing data
print(df.isnull().sum())
df=df.fillna(0)
```

```python
#4 selecting and filtering data
ages=df[df['age']>25]
print(ages)
```

```python
#5 adding and removing data
df['total']=df['a']+df['b']
print(df.head())
```

```python
#6 grouping data(groupby)
res=df.groupby('category')['sales'].sum()
print(res)
```
