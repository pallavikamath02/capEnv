```python
#bar plot
fruit=['apples','mangoes','bananas','cherries','dates']
sales=[400,500,350,300,450]
plt.bar(fruit,sales)
plt.xlabel("Fruit")
plt.ylabel("Sales")
plt.title("Fruit Sales bar plot")
plt.show()
```

```python
#horizontal bar plot
fruit=['apples','mangoes','bananas','cherries','dates']
sales=[400,500,350,300,450]
plt.barh(fruit,sales)
plt.xlabel("Sales")
plt.ylabel("Fruit")
plt.title("Fruit Sales-horizontal bar plot")
plt.show()
```

```python
#stacked bar plot
n=5
b=[20,35,30,35,27]
g=[25,32,20,25,35]
bs=(2,3,4,1,2)
gs=(3,5,2,3,3)
ind=np.arange(n)
width=0.35
fig=plt.subplots(figsize=(10,7))
p1=plt.bar(ind,b,width,yerr=bs)
p2=plt.bar(ind,g,width,bottom=b,yerr=gs)
plt.title("Contribution by teams")
fig[1].set_ylabel("Contribution")
plt.xticks(ind,('T1','T2','T3','T4','T5'))
plt.yticks(np.arange(0,81,10))
plt.legend((p1[0],p2[0]),('boys','girls'))
plt.show()
```

```python
#customized histogram with watermark
from matplotlib import colors
np.random.seed(2368752)
n_p=1000
n_b=30
x=np.random.randn(n_p)
y=0.8**x+np.random.randn(n_p)+25
legend=['Distribution']
fig,axs=plt.subplots(1,1,figsize=(10,7),tight_layout=True)
for s in ['top','bottom','left','right']:
    axs.spines[s].set_visible(False)
axs.xaxis.set_ticks_position('none')
axs.yaxis.set_ticks_position('none')
axs.xaxis.set_tick_params(pad=5)
axs.yaxis.set_tick_params(pad=10)
axs.grid(visible=True,color='grey',linestyle='-',linewidth=0.5,alpha=0.6)
fig.text(0.9,0.15,'Jeeteshgavande30',fontsize=12,color='red',ha='right',va='bottom',alpha=0.7)
N,bins,patches=axs.hist(x,bins=n_b)
fracs=((N**(1/5))/N.max())
norm=colors.Normalize(fracs.min(),fracs.max())
for thisfrac,thispatch in zip(fracs,patches):
    color=plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)
plt.xlabel("X-Axis")
axs.set_ylabel("Y-Axis")
plt.legend(legend)
plt.title("customized histogram with watermark")
plt.show()
```

```python
#python seaborn tutorial
import seaborn as sns
sns.set(style="white")
rs=np.random.RandomState(10)
d=rs.normal(size=100)
sns.histplot(d,kde=True,color='m')
```

```python
#lineplot
sns.set(style="dark")
fmri=sns.load_dataset("fmri")
sns.lineplot(x='timepoint',y='signal', hue='region',style='event',data=fmri)
plt.show()
```

```python
#strip plot
x=['mon','tues','wed','thurs','fri','sat','sun']
y=[6.7,4,6,2,4.9,1.8,5]
ax=sns.stripplot(x=x,y=y)
ax.set(xlabel='Days',ylabel='Ammount Spent')
plt.title("Daily Spending (Custom Data)")
plt.show()
```

```python
#swarm plot
sns.set(style="whitegrid")
iris=sns.load_dataset("iris")
sns.swarmplot(x="species",y="sepal_length",data=iris)
plt.title("Swarm plot of sepal length by species")
plt.show()
```

```python
#count plot
tips=sns.load_dataset("tips")
print(tips.head())
sns.countplot(x="sex",data=tips)
sns.countplot(x="day",hue="smoker",data=tips)
plt.show()
```

```python
#box plot
tips=sns.load_dataset("tips")
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
plt.title("Total bill distribution by day and smoking data")
plt.show()
```

```python
#violin plot
tips=sns.load_dataset("tips")
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
plt.title("Violin Plot of Total bill by day and gender")
plt.show()
```

```python
#strip plot with hue
tips=sns.load_dataset("tips")
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='smoker',dodge=True)
plt.title("Total bill distribution with smoking status")
plt.show()
```

```python
#enhancing seaborn histogram with matplotlib annotations
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=np.random.randn(1000)
plt.figure(figsize=(8,5))
sns.histplot(data, kde=True, bins=30, color='skyblue')

#adding mean line using matplotlib
mean_value = np.mean(data)
plt.axvline(mean_value, color='red', linestyle='--',linewidth=2, label='Mean')
plt.text(mean_value + 0.1, 50, 'Mean: {:.2f}'.format(mean_value), color='red')
plt.title("Disturbution with seaboarnand matplotlib Histogram with Mean Line")
plt.xlabel("Value")
plt.ylabel("Frequency")     
plt.legend()
plt.show()
```
