```python
#use of with closes file automatically after the block
with open("sample.txt","r") as f:
    content=f.read()
    print(content)
```

```python
#for loop to read line by line
with open("sample.txt","r") as f:
    for l in f:
        print(l.strip())
```

```python
#readline() used to read 1 line at a time inside loop
with open("sample.txt","r") as f:
    l=f.readline()
    while l:
        print(l.strip())
        l=f.readline()
```

```python
#overwriting a file
with open("sample.txt","w",encoding="utf-8") as f:
    f.write("Edited with write mode!!\n")
    f.write("Another line\n")
with open("sample.txt","r") as f:
    print(f.read())
```

```python
#append to existing file
with open("sample.txt","a",encoding="utf-8") as f:
    f.write("Hello World\n")
    f.write("Appended Lines")
with open("sample.txt","r") as f:
    print(f.read())
```

```python
#create if file doesnt exist
try:
    with open("sample1.txt","x",encoding="utf-8") as f:
        f.write("Created using exclusive mode!!")
except FileExistsError:
    print("sample1.txt already exists, exclusive creation aborted!!")
f=open("sample1.txt","r")
print(f.read())
```

```python
l=["Line1","Line2","Line3"]
t="\n".join(l)+"\n"
with open("sample1.txt","w",encoding="utf-8") as f:
    f.write(t)
with open("sample1.txt","r") as f:
    print(f.read())
```

```python
#writing to binary file
data=b'\x00\x01\x02\x03\x04'
with open("example.bin", "wb") as bf:
    bf.write(data)
with open("example.bin", "rb") as bf:
    print(bf.read())
```

```python
#handling exceptions while closing file
try:
    f=open("sample.txt","r") 
    c=f.read()
    print(c)
finally:
    f.close()
```

```python
try:
    with open("bank.log", "w",encoding="utf-8") as f:
        f.write("Transaction Success!!")
    print("Transaction Logged!!")
except Exception:
    print("Logging Error!!")
```

```python
try:
    bal=int(input("Enetr the balance: "))
    with open("bank.log","w",encoding="utf-8") as f:
        if bal==0:
            f.write("Account Closed!!")
        else:
            f.write("Account Has pending Banlance!!")
    print("Status Updated!!")
except Exception:
    print("Logging Error!!")
```
