```python
#Tower of Hanoi
def hanoi(n, source, target, auxiliary):
    if n <= 0:
        return
    hanoi(n-1, source, auxiliary, target)
    print(f"Move disk {n} from {source} to {target}")
    hanoi(n-1, auxiliary, target, source)

n = int(input("Enter number of disks (less than 3): "))
print(f"Solving Tower of Hanoi with {n} disks")
hanoi(n, 'A', 'C', 'B')
```

```python
#print hourglass pattern using asterisks
def print_hourglass(n: int) -> None:
    if n <= 0:
        return
    for i in range(n):
        indent = ' ' * i
        stars = ' '.join('*' for _ in range(n - i))
        print(indent + stars)
    for k in range(n - 2, -1, -1):
        indent = ' ' * k
        stars = ' '.join('*' for _ in range(n - k))
        print(indent + stars)
n = int(input("Enter a number: ").strip())
print_hourglass(n)
```

```python
#print the reversed word order of the input string
def reverse_words(s):
    words = []
    word = ""
    for ch in s:
        if ch == " ":
            words.append(word)
            word = ""
        else:
            word += ch
    words.append(word)  
    rs = ""
    for i in range(len(words) - 1, -1, -1):
        rs += words[i]
        if i != 0:
            rs += " "
    return rs
s = "god is great"
print(reverse_words(s))
```
