```python
#python exception handling
try:
    n=int(input("Enter a number: "))
    res=100/n
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Error: Invalid value encountered.")
else:
    print("Result is:", res)
finally:
    print("Execution completed.")
```

```python
#raiseing error for exception handling
class AgeError(Exception):
    pass
def set(age):
    if age<0:
        raise AgeError("Age cannot be negative")
    print(f"Age set to {age}")
try:
    set(-5)
except AgeError as e:
    print("Error:",e)
```

```python
#Daily withdrawal limit
class DailylimitError(Exception):
    pass
def withdraw(amount):
    if amount>25000:
        raise DailylimitError("Daily Limit Exceeded")
    print(f"Withdrew {amount} successfully")
try:
    withdraw(int(input("Enter withdrawal amount: ")))
except DailylimitError as e:
    print("Error:",e)
```

```python
#pin verification system
class PinError(Exception):
    pass
def verify(pin):
    if pin!="1234":
        raise PinError("Invalid PIN!!")
    print("Access Granted!!")
try:
    verify(input("Enter the PIN: "))
except PinError as e:
    print("Error:",e)
```

```python
#loan eligibility check
class LoanError(Exception):
    pass
def eligibility(sal):
    if sal>=20000:
        print("Loan Approved!!")
    else:
        raise LoanError("Loan Rejected!!")
try:
    eligibility(int(input("Enter the Salary: ")))
except LoanError as e:
    print("Error: ",e)
```

```python
#multistage baning management system
user=input("Enter Username: ")
pwd=input("Enter Password: ")
bal=int(input("Enter initial balance: "))
wa=int(input("Enter withdrawal amount: "))
try:
    if user=="user1" and pwd=="pass123":
        if wa<=bal:
            bal-=wa
            print("Transaction Successful!!")
            print("Remaining Balance:",bal)
        else:
            print("Insufficient Balance!!")
    else:print("Authentication Failed!!")
except ValueError:
    print("Authentication Failed!!")
finally:
    print("Session Closed.")
    
```

```python
#nested try-except online payment gateway

try:
    bal=int(input("Enter payment amount: "))
    card=input("Enter card number: ")
    pa=int(input("Enter payment amount: "))
    try:
        if len(card)!=16 or not card.isdigit():
            raise ValueError("Invalid Card Number")
        if pa>5000:
            raise Exception("Amount exceeds limit")
        if bal<pa:
            raise Exception("Insufficient Balance")
        print("Payment Successful!!")
    except ValueError as ve:
        print("Card Error:",ve)
except Exception as e:
    print("Payment Failed:",e)
```

```python
#custom loan system
class LoanError(Exception):
    pass    
def apply_loan(sal,cs):
    if sal<25000:
        print("Low salary!!")
        raise LoanError("Loan Rejected!!")
    if cs<700:
        print("Low credit score!!")
        raise LoanError("Loan Rejected!!")
    print("Loan Approved")
try:
    sal=int(input("Enter Salary: "))
    cs=int(input("Enter Credit Score: "))
    apply_loan(sal,cs)
except LoanError as le:
    print("Error:",le)
```
